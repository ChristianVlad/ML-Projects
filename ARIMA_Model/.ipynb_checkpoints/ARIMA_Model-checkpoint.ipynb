```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
from statsmodels.tsa.stattools import adfuller

pd.options.display.float_format = '{:.2f}'.format

file = 'data_arima.csv'

data = pd.read_csv(file, sep=',')
```

```python
data
```

```python
# Calcular el primer y tercer cuartil (Q1 y Q3)
Q1 = data['deposito'].quantile(0.25)
Q3 = data['deposito'].quantile(0.75)

# Calcular el rango intercuartílico (IQR)
IQR = Q3 - Q1

# Definir los límites superior e inferior para detectar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar los valores que están dentro del rango
data = data[(data['deposito'] > lower_bound) & (data['deposito'] < upper_bound)]
```

```python
# Asegúrate de que la columna de fechas esté en formato datetime
data['date_tx'] = pd.to_datetime(data['date_tx'])

# Establece la columna de fechas como el índice del DataFrame
data.set_index('date_tx', inplace=True)

start_date = datetime(2023,6,3)
end_date = datetime(2024,4,11)
data = data[start_date:end_date]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['deposito'])
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.title('Visualización de los datos de la serie temporal')
plt.show()
```

```python
def test_stationary(dataFrame, var):
    dataFrame['rollingMean']= dataFrame[var].rolling(7).mean()
    dataFrame['rollingStd']= dataFrame[var].rolling(7).std()

    from statsmodels.tsa.stattools import adfuller
    
    adfuller = adfuller(dataFrame[var], autolag='AIC')

    stats = pd.Series(adfuller[0:4], index=['Test Statistic:', 'p-value:', '# Lags Used:', 'Number of Observations Used:'])
    print(stats)
    print('---------------------------------------')

    for key, values in adfuller[4].items():
        print('criticality', key,":", values)

    plt.figure(figsize=(15, 5))
    plt.plot(dataFrame.index, dataFrame[var], label=var)
    plt.plot(dataFrame.index, dataFrame['rollingMean'], label='Rolling Mean')
    plt.plot(dataFrame.index, dataFrame['rollingStd'], label='Rolling Std')
    
    # Adding legend
    plt.legend()
    plt.show()
```

```python
test_stationary(data, 'deposito')
```
